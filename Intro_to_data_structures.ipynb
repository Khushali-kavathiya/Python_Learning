From ndarray



If data is an ndarray, index must be the same length as data. If no index is passed, one will be created having values [0, ..., len(data) - 1].



```python
s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
```

**From dict**


```python
d = {"b": 1, "a": 0, "c": 2}
```

```python
pd.Series(d)
```

If an index is passed, the values in data corresponding to the labels in the index will be pulled out.



```python
d = {"a": 0.0, "b": 1.0, "c": 2.0}
```

```python
pd.Series(d)
```

```python
pd.Series(d, index=["b", "c", "d", "a"])
```

If data is a scalar value, the value will be repeated to match the length of index. If the index is not provided, it defaults to RangeIndex(1).



```python
pd.Series(5.0, index=["a", "b", "c", "d", "e"])
```

# Series is ndarray-like

Series acts very similarly to a numpy.ndarray and is a valid argument to most NumPy functions. However, operations such as slicing will also slice the index.



```python
s.iloc[0]
```

```python
s.iloc[:3]
```

```python
s[s > s.median()]
```

```python
s.iloc[[4,3,1]]
```

```python
np.exp(s)
```

```python
s.dtype
```

```python
s.array
```

```python
s.to_numpy()
```

# Series is dict-like

A Series is also like a fixed-size dict in that you can get and set values by index label:



```python
s["a"]
```

```python
s["e"]
```

```python
s
```

Using the Series.get() method, a missing label will return None or specified default:



```python
s.get("f")
```

 # From structured or record array

```python
data = np.zeros((2,), dtype=[("A", "i4"), ("B", "f4"), ("C", "S10")])
```

```python
data[:] = [(1, 2.0, "Hello"), (2, 3.0, "World")]
```

```python
pd.DataFrame(data)
```

# From a list of namedtuples

```python
from collections import namedtuple
```

```python
Point = namedtuple("Point", "x y")
```

```python
pd.DataFrame([Point(0, 0), Point(0, 3), (2, 3)])
```

```python
Point3D = namedtuple("Point3D", "x y z")
```

```python
pd.DataFrame([Point3D(0,0,0), Point3D(0,3,5), Point(2,3)])
```
