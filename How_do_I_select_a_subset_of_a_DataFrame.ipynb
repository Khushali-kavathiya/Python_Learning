### How to I filter specific rows from a DataFrame?

I'm interested in the passengers older than 35 years

```python
above_35 = titanic[titanic['Age'] > 35]
```

```python
above_35
```

```python
titanic['Age'] > 35
```

```python
above_35.shape
```

I'm inetrested in the titanic passengers from cabin class 2 and 3

```python
class_23 = titanic[titanic['Pclass'].isin([2,3])]
```

```python
class_23
```

```python
class_23.shape
```

```python
titanic['Pclass'].isin([2,3])
```

Using the |(or) operator

```python
class_23 = titanic[(titanic['Pclass'] == 2) | (titanic['Pclass'] == 3)]
```

```python
class_23.head()
```

I want to work with passenger data for which age is know

```python
age_no_na = titanic[titanic['Age'].notna()]
```

```python
age_no_na.head()
```

```python
titanic['Age'].notna()
```

```python
age_no_na.shape
```
