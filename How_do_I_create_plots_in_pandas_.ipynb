```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv('air_quality_no2.csv', index_col=0, parse_dates=True)
```

I want a quick visual check of the data.

```python
air_quality.plot()
```

I want to plot only the column of the data table with the data from Paris.

```python
air_quality.head()
```

```python
air_quality['station_paris']
```

```python
type(air_quality['station_paris'])
```

```python
air_quality['station_paris'].plot()
```

I want to visually compare the NO2 values measured in London versus Paris.

```python
air_quality.plot.scatter(x='station_london', y='station_paris', alpha=0.5)
```

```python
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
air_quality.plot.box()
```

I want to each of the column in a separete subplot.

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
```

I want to further customize, extend or save the resulting plot.


```python
fig, axs = plt.subplots(figsize=(12, 4))

```

```python
air_quality.plot.area(ax=axs)
```

```python
axs.set_ylabel("NO$_2$ concentration")
```

```python
fig.savefig("no_concentration.png")
```

```python
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))
air_quality.plot.area(ax=axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("no2_concentration2.png")
plt.show()
```
